```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from markdown2 import Markdown
from openai_tools import call_with_sys
markdowner = Markdown()
df = pd.read_csv("ARK-1.csv")
```

```python
import os

def comp_info(comp_url):
    html_text = requests.get(comp_url).text
    soup = BeautifulSoup(html_text, 'html.parser')
    article = soup.find(id='main').article
    comp_desc = article.header.text
    single_content = article.select(".single-content")
    # abstract = single_content[0].select(".abstract")[0].text
    ctt = single_content[0].text
    return comp_desc, ctt

def md_comp_info(ticker):
    url_t=f"https://www.mg21.com/{ticker}.html"
    comp_desc, ctt = comp_info(url_t)
    
    with open(f"{ticker}.html", 'w') as f:
        md = f"## {comp_desc} \n\n{ctt}"
        html = markdowner.convert(md)
        f.write(html)
    
    sys_content = "针对用户给出的文章，提取200字的中文摘要。"
    cid, abstr = call_with_sys(sys_content, ctt, max_tokens=1000)

    with open(f"{ticker}_ABST.html", 'w') as f:
        md = f"## {comp_desc} \n\n{abstr}"
        html = markdowner.convert(md)
        f.write(html)

    return comp_desc

# md_comp_info(df['Ticker'][2])

def all_comp():
    all_desc = []
    for ticker in df['Ticker']:
        url_t=f"https://www.mg21.com/{ticker}.html"
        try:
            file_path = f"{ticker}_raw.html"
            html_text = requests.get(url_t).text
            soup = BeautifulSoup(html_text, 'html.parser')
            article = soup.find(id='main').article
            comp_desc = article.header.text
            print(comp_desc)
            all_desc.append(comp_desc)
            if os.path.exists(file_path):
                continue
            with open(file_path, 'w') as f:
                f.write(html_text)
        except:
            print('error', url_t)
            continue
    print(all_desc)

all_comp()
```
